## –ø—Ä–æ—Å—Ç–æ–π –∑–∞–ø—Ä–æ—Å

```python
import openai

YANDEX_CLOUD_FOLDER = ""
YANDEX_CLOUD_API_KEY = ""
YANDEX_CLOUD_MODEL = "aliceai-llm/latest"


client = openai.OpenAI(
    api_key=YANDEX_CLOUD_API_KEY,
    base_url="https://rest-assistant.api.cloud.yandex.net/v1",
    project=YANDEX_CLOUD_FOLDER
)

response = client.responses.create(
    model=f"gpt://{YANDEX_CLOUD_FOLDER}/{YANDEX_CLOUD_MODEL}",
    temperature=0.94,
    instructions="—Ç—ã —Å—É–ø–µ—Ä –∫—Ä–µ–∞—Ç–∏–≤–Ω—ã–π –º–∞—Ä–∫–µ—Ç–æ–ª–æ–≥",
    input="–ø—Ä–∏–≤–µ—Ç, –º—ã —Å–µ–≥–æ–¥–Ω—è –≤–µ–¥–µ–º –ª–µ–∫—Ü–∏—é –Ω–∞ —Ç–µ–º—É –ò–ò –∞–≥–µ–Ω—Ç–æ–≤ –∏ –∏–Ω—Ç–µ–≥—Ä–∞—Ü–∏–∏ –∏—Ö –≤ –ò–¢ –ª–∞–Ω–¥—à–∞—Ñ—Ç",
    max_output_tokens=500
)

print(response.output_text)
```

## –ø—Ä–æ—Å—Ç–æ–π –∑–∞–ø—Ä–æ—Å

```python
# Please install OpenAI SDK first: `pip3 install openai`
import os
from openai import OpenAI

DEEPSEEK_API_KEY = ''

client = OpenAI(api_key=DEEPSEEK_API_KEY, base_url="https://api.deepseek.com")

response = client.chat.completions.create(
    model="deepseek-chat",
    messages=[
        {"role": "system",
         "content": "–ü—Ä–∏–≤–µ—Ç, —Ç—ã –º–æ–π –¥—Ä—É–≥ –ø–∏—à–∏ –Ω–∞ —Ä—É—Å—Å–∫–æ–º —è–∑—ã–∫–µ"},
        {"role": "user",
         "content": "–ü—Ä–∏–≤–µ—Ç, –∫–∞–∫ –¥–µ–ª–∞"},
    ],
    stream=False
)

print(response.choices[0].message.content)
```

```python
promt = '''–¢—ã —Å—É–ø–µ—Ä –∞–Ω–∞–ª–∏—Ç–∏–∫ —Ç–≤–æ—è –∑–∞–∞—á–∞ –∞–Ω–∞–ª–∏–∑–∏—Ä–æ–≤–∞—Ç—å –±–∞–Ω–∫–æ–≤—Å–∫–∏–µ –¥–æ–∫—É–º–µ–Ω—Ç—ã –∏ –≤—ã–¥–∞–≤–∞—Ç—å —Å–≤–æ–¥—ã–Ω–π –æ—Ç—á–µ—Ç,
            —á—Ç–æ–±—ã —É–±–µ–¥–∏—Ç—å—Å—è, —á—Ç–æ –¥–∞–Ω–Ω—ã–µ –æ –∑–∞–µ–º—â–∏–∫–∞—Ö –≤–µ—Ä–Ω—ã–µ,
            —Ç–∞–∫–∂–µ —Ç—ã –≥–µ–Ω–µ—Ä–∏—Ä—É–µ—â—à—å –æ—Ç—á–µ—Ç –≤ –ª–µ–¥—É—é—â–µ–º –≤–∏–¥–µ:
            1. –æ —á–µ–º —ç—Ç–∏ –¥–∞–Ω–Ω—ã–µ?
            2. –∫–∞–∫–∏–µ –æ—Å–Ω–æ–≤–Ω—ã–µ –∑–∞–∫–æ–Ω–æ–º–µ–Ω—Ä–Ω–æ—Å—Ç–∏ —Ç—ã –≤–∏–¥–∏—à—å?
            3. –∫–∞–∫–∏–µ –≤—ã–≤–æ–¥—ã –∏ —Ä–µ–∫–æ–º–µ–Ω–¥–∞—Ü–∏–∏ —Ç—ã –º–æ–∂–µ—à—å —Å–¥–µ–ª–∞—Ç—å –æ—Ç–Ω–æ—Å–∏—Ç–µ–ª—å–Ω–æ –∞–Ω–∞–ª–∏–∑–∞ –¥–∞–Ω–Ω—ã—Ö'''
```

```python
response = client.chat.completions.create(
    model="deepseek-chat",
    messages=[
        {"role": "system",
         "content": promt },
        {"role": "user",
         "content": file_content[:15000]},
    ],
    stream=False
)

print(response.choices[0].message.content)
```
